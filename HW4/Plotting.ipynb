```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
val = []
train = []
with open("torch-rnn/log_training_songs.txt","r") as f:
    i = 0
    for line in f:
        i += 1
        if i == 1:
            continue
        words = line.strip().split()
        if words[0] == "val_loss":
            val.append((epoch,float(words[-1])))
        else:
            epoch = float(words[1])
            loss = float(words[-1])
            train.append((epoch,loss))
        
```

```python
import numpy as np
```

```python
train = np.array(train)[np.sort(np.random.choice(len(train), 100))]
```

```python
sns.set_palette("husl")
```

```python
x1 = [i[0] for i in train]
x2 = [i[0] for i in val]
y1 = [i[1] for i in train]
y2 = [i[1] for i in val]

# sns.plt.ylim([91,100])
sns.plt.xlabel("Epochs")
sns.plt.ylabel("Train/Val loss")
sns.plt.plot(x1, y1,  label = "Train loss" )
sns.plt.plot(x2, y2,  label = "Val loss" )
sns.plt.title("LSTM training for Music composition")
sns.plt.legend(loc='upper left')
```
